```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# 1. CRIAÇÃO DO GRAFO
# Vamos criar um grafo com 6 vértices ('A', 'B', 'C', 'D', 'E', 'F')
# Vertices são os pontos do grafo. (Quando pensamos em um mapa, os vértices podem ser cidades)
# Quando eles são direcionados, podemos chamar de nós. (Muito comum em árvores e redes neurais)
G = nx.Graph()
vertices = ['A', 'B', 'C', 'D', 'E', 'F']
G.add_nodes_from(vertices)
```

```python
# Adiciona arestas entre os vértices. (Arestas são conexões entre os vértices)
# Arestas que formam um ciclo A-B-C-F-E-D-A
# Adiciona algumas arestas extras para aumentar a complexidade.
arestas = [('A', 'B'), ('B', 'C'), ('C', 'F'), ('F', 'E'), ('E', 'D'), ('D', 'A'),('A', 'C'), ('B', 'D'), ('C', 'E')]
G.add_edges_from(arestas)
```

```python
# 2. DEFINIÇÃO DOS CAMINHOS PARA ILUSTRAR
# O caminho válido (O Circuito Hamiltoniano)
caminho_valido = ["A", "B", "C", "F", "E", "D", "A"]
caminho_valido2 = ["A", "B", "D", "E", "F", "C", "A"]
caminho_valido3 = ["A", "D", "E" , "F", "C", "B", "A"]
# Um caminho inválido (não visita todos os vértices)
caminho_invalido = ["A", "C", "E", "F"]
```

```python
# 3. PREPARAÇÃO PARA VISUALIZAÇÃO
# Converte os caminhos em listas de arestas para o networkx desenhar
arestas_validas1 = list(zip(caminho_valido, caminho_valido[1:]))
arestas_validas2 = list(zip(caminho_valido2, caminho_valido2[1:]))
arestas_validas3 = list(zip(caminho_valido3, caminho_valido3[1:]))
arestas_invalidas = list(zip(caminho_invalido, caminho_invalido[1:]))


# Define uma posição fixa para os nós, para que a imagem não mude a cada execução
pos = nx.spring_layout(G, seed=42)
```

```python
plt.figure(figsize=(8, 6))
plt.title("Grafo de Exemplo", fontsize=14)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12)
```

```python
# 4. DESENHANDO O GRAFO E OS CAMINHOS (Forma 1)
plt.figure(figsize=(8, 6))

# Subplot 1: Mostrando o Circuito Hamiltoniano
plt.title("Exemplo-1 de Circuito Hamiltoniano (Válido)", fontsize=14)
plt.text(0.5, 0.98, "Caminho Válido: " + " -> ".join(caminho_valido), horizontalalignment='center', transform=plt.gca().transAxes, fontsize=10)
# Desenha o grafo base
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12)
# Desenha as arestas do circuito em verde
nx.draw_networkx_edges(G, pos, edgelist=arestas_validas1, edge_color='green', width=4.5)
```

```python
# 4. DESENHANDO O GRAFO E OS CAMINHOS (Forma 1)
plt.figure(figsize=(8, 6))

# Subplot 1: Mostrando o Circuito Hamiltoniano
plt.title("Exemplo-2 de Circuito Hamiltoniano (Válido)", fontsize=14)
plt.text(0.5, 0.98, "Caminho Válido: " + " -> ".join(caminho_valido2), horizontalalignment='center', transform=plt.gca().transAxes, fontsize=10)
# Desenha o grafo base
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=800, font_size=12)
# Desenha as arestas do circuito em verde
nx.draw_networkx_edges(G, pos, edgelist=arestas_validas2, edge_color='purple', width=4.5)
```

```python
# Subplot 2: Mostrando um caminho Inválido
plt.figsize=(8, 6)
plt.title("Exemplo de Caminho Não Hamiltoniano (Inválido)", fontsize=14)
# Desenha o grafo base
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=12)
# Desenha as arestas do caminho inválido em vermelho
nx.draw_networkx_edges(G, pos, edgelist=arestas_invalidas, edge_color='red', width=4.5, style='dashed')

plt.show()
```

```python
# 4. DESENHANDO O GRAFO E OS CAMINHOS
plt.figure(figsize=(18, 6))
plt.suptitle("Demonstração de Caminhos Hamiltonianos Válidos", fontsize=16)

# Subplot 1: Mostrando o Circuito Hamiltoniano
plt.subplot(1, 3, 1)
plt.title("Exemplo-1 de Circuito Hamiltoniano (Válido)", fontsize=14)
plt.text(0.5, 0.98, "Caminho Válido: " + " -> ".join(caminho_valido), horizontalalignment='center', transform=plt.gca().transAxes, fontsize=10)
# Desenha o grafo base
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=12)
# Desenha as arestas do circuito em verde
nx.draw_networkx_edges(G, pos, edgelist=arestas_validas1, edge_color='green', width=4.5)

# Subplot 2: Mostrando outro caminho valido
plt.subplot(1, 3, 2)
plt.title("Exemplo-2 de Caminho Hamiltoniano (Válido)", fontsize=14)
plt.text(0.5, 0.98, "Caminho Válido: " + " -> ".join(caminho_valido2), horizontalalignment='center', transform=plt.gca().transAxes, fontsize=10)
# Desenha o grafo base
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=12)
# Desenha as arestas do caminho válido em roxo
nx.draw_networkx_edges(G, pos, edgelist=arestas_validas2, edge_color='purple', width=4.5)

# Subplot 3: Mostrando o Segundo Circuito Hamiltoniano
# plt.subplot(1, 3, 3)
# plt.title("Exemplo-3 de Caminho Hamiltoniano (Válido)", fontsize=14)
# # Desenha o grafo base
# nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=12)
# # Desenha as arestas do caminho válido em laranja
# nx.draw_networkx_edges(G, pos, edgelist=arestas_validas3, edge_color='orange', width=2.5)

# Subplot 3: Mostrando um caminho Inválido
plt.subplot(1, 3, 3)
plt.figsize=(8, 6)
plt.title("Exemplo de Caminho Não Hamiltoniano (Inválido)", fontsize=14)
# Desenha o grafo base
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=12)
# Desenha as arestas do caminho inválido em vermelho
nx.draw_networkx_edges(G, pos, edgelist=arestas_invalidas, edge_color='red', width=4.5, style='dashed')

plt.tight_layout()
plt.show()
```

### Exemplo de circuito hamiltoniano no networkx

```python
G1 = nx.complete_graph(range(3))
G2 = nx.complete_graph(range(4))
G3 = nx.complete_graph(range(5))
G12 = nx.disjoint_union(G1, G2)
G = nx.disjoint_union(G12, G3)
nx.add_path(G, [0,3])
nx.add_path(G, [4,8])
nx.add_path(G, [1,7])
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1,3,1)
nx.draw(G1, with_labels=True)

plt.subplot(1,3,2)
nx.draw(G2, with_labels=True)

plt.subplot(1,3,3)
nx.draw(G3, with_labels=True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4, 4))
nx.draw(G, with_labels = True)
```

```python
CycleList = sorted(nx.simple_cycles(G))
HamiltonianCycles = 0
for list in CycleList:
    if len(list) == len(G.nodes):
        HamiltonianCycles += 1
        print("Ciclo Hamiltoniano encontrado: ", list)
print("Número total de ciclos Hamiltonianos: ", HamiltonianCycles)
```
